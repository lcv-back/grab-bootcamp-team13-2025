````python
import os
import shutil
from pathlib import Path

def move_all_images(src_dir: str, dest_dir: str):
    """
    Chuyển tất cả các file ảnh từ src_dir sang dest_dir.
    
    Args:
        src_dir (str): đường dẫn tới thư mục nguồn.
        dest_dir (str): đường dẫn tới thư mục đích.
    """
    src_path = Path(src_dir)
    dest_path = Path(dest_dir)
    
    # Tạo thư mục đích nếu chưa tồn tại
    dest_path.mkdir(parents=True, exist_ok=True)
    
    # Các phần mở rộng file ảnh cần chuyển
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff'}
    
    for root, _, files in os.walk(src_path):
        for filename in files:
            file_ext = Path(filename).suffix.lower()
            if file_ext in image_extensions:
                src_file  = Path(root) / filename
                dest_file = dest_path       / filename

                # Nếu file đích đã tồn tại, xóa nó để ghi đè
                if dest_file.exists():
                    dest_file.unlink()  # xóa file cũ :contentReference[oaicite:3]{index=3}

                # Thực hiện di chuyển (ghi đè) :contentReference[oaicite:4]{index=4}
                shutil.move(str(src_file), str(dest_file))
                print(f"Moved: {src_file} -> {dest_file}")
````

````python
check_dir = "./dataunzip"
data_dir = "./dataunzip/data"



for list_dir in os.listdir(check_dir):
    list_dir = os.path.join(check_dir, list_dir)
    for dir in os.listdir(list_dir):
        # dir = os.path.join(list_dir, dir)
        src_dir = os.path.join(list_dir, dir)
        dest_dir = os.path.join(data_dir, dir)
        print(src_dir, dest_dir)
        move_all_images(src_dir, dest_dir)
 
````

````python
import re
def parse_symptoms_raw(raw: str):
    raw = re.sub(r"^```(?:json)?\s*", "", raw.strip())
    raw = raw.replace("```", "").strip()
    
    start = raw.find('[')
    end   = raw.rfind(']')
    if start == -1 or end == -1:
        raise ValueError("Không tìm thấy dấu [ hoặc ] trong symptoms_raw")
    json_text = raw[start:end+1]
    
    return json.loads(json_text)
````

````python
from PIL import Image
from tqdm import tqdm
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def add_image_to_database(self, input_data=[], check_existing_path=False,
                          batch_size=100, sim_threshold=0.99):

    if check_existing_path:
        existing_paths = set(self.database1.to_pandas()['path'].tolist())

    batch = []
    batch_embs = []
    batch_paths = []

    for name, path in tqdm(input_data, desc="image process"):
        if check_existing_path and path in existing_paths:
            print(f"Path already exists in database, skipping: {path}")
            continue

        image = Image.open(path)
        inputs = self.processor(images=image, return_tensors="pt").to(self.device)
        with torch.no_grad():
            emb = (self.model
                   .get_image_features(**inputs)
                   .cpu()
                   .squeeze()
                   .numpy()
                   .reshape(1, -1))

        results = self.database1.search(image_embedding, vector_column_name='image_embedding').metric("cosine").limit(1).to_list()

        if results != []:
            similar = 1 - results[0].get('_distance', None)
            if similar > 0.99:
                similar_path = results[0].get('path')
                print(f"Embedding like {similar_path} ({similar:.3f}), skip: {path}")
                continue

        if batch_embs:
            arr = np.vstack(batch_embs)
            sims = cosine_similarity(arr, emb)
            max_sim = sims.max()
            if max_sim > sim_threshold:

                idx_sim = sims.argmax()
                dup_path = batch_paths[idx_sim]
                print(f"[Batch] Too similar to {dup_path} (sim={max_sim:.3f}), skipping: {path}")
                continue

        batch.append({
            "image_embedding": emb.flatten(),
            "name": name,
            "path": path
        })
        batch_embs.append(emb)
        batch_paths.append(path)
        if check_existing_path:
            existing_paths.add(path)

        if len(batch) >= batch_size:
            self.database1.add(batch)
            print(f"Flushed batch of {len(batch)} items to database.")
            batch.clear()
            batch_embs.clear()
            batch_paths.clear()

    if batch:
        self.database1.add(batch)
        print(f"Flushed final batch of {len(batch)} items to database.")

    print("Done processing all images.")
````

````python
def compare_line_by_line(path1, path2):
    with open(path1, 'r', encoding='utf-8') as f1, \
         open(path2, 'r', encoding='utf-8') as f2:
        for i, (l1, l2) in enumerate(zip(f1, f2), start=1):
            if l1.rstrip() != l2.rstrip():
                print(f"Dòng {i} khác nhau:")
                print(f"  file1: {l1!r}")
                print(f"  file2: {l2!r}")

# Ví dụ gọi
compare_line_by_line('symptom disease.json', 'fixed_selected.json')
````

````python
import json
from google import genai

# Load the list of diseases and their symptoms
with open("symptom disease.json", "r", encoding="utf-8") as f:
    symptom_list = json.load(f)

# Initialize GenAI client
private_key = "AIzaSyAaKsU8QFAqFmkvoG9O3KnbYcXR6cBsd5c"
client = genai.Client(api_key=private_key)

# Prepare data structure to hold rewritten symptoms
data = []

for item in symptom_list:
    disease = item.get("disease_name")
    original_symptoms = [s.get("symptom_name") for s in item.get("symptoms", [])]

    # Build prompt to rewrite symptom names
    formatted_list = ", ".join(original_symptoms)
    prompt = (
        f"Below is a list of medical symptom names: {formatted_list}."
        " Please rewrite each symptom using an alternative English term or synonym,"  
        " and return the list only as a JSON array of strings in the same order."
    )

    # Call Gemini to generate rewritten symptom names
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=[prompt]
    )

    # Parse Gemini's JSON array of rewritten symptoms
    try:
        rewritten = json.loads(response.text.strip())
    except json.JSONDecodeError:
        # Fallback: split by commas if JSON fails
        rewritten = [s.strip() for s in response.text.split(",")]

    # Append to output list
    data.append({
        "disease_name": disease,
        "symptom_name": rewritten
    })

# Print results in JSON format
print(json.dumps(data, ensure_ascii=False, indent=2))
````
