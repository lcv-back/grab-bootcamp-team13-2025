```python
# ------------ CONFIG ------------
# 1) Lấy API key từ Kaggle Secrets
API_KEY = UserSecretsClient().get_secret("anhquancho")

# 2) Đường dẫn file CSV bạn vừa gắn (ví dụ thư mục '/kaggle/input/diseases-clean/')
CSV_PATH   = "/kaggle/input/dataset-clean/diseases_details_clean.csv"
# 3) File output (luôn ghi vào /kaggle/working để Kaggle cho phép tải về)
OUTPUT_JSONL = "/kaggle/working/diseases_test2470.jsonl"

MODEL_NAME   = "gemini-2.0-flash"
REQS_PER_MIN = 2000
SLEEP_SEC    = 60.0 / REQS_PER_MIN
CHKP_EVERY   = 100         # flush sau N bản ghi
RESUME_RUN   = True       # bật resume

# ------------ INIT ------------
genai_client = genai.Client(api_key=API_KEY)
df = pd.read_csv(CSV_PATH, encoding='cp1252', encoding_errors='replace')
print("Loaded:", df.shape)
```

```python
def merge_row(row) -> str:
    """
    Ghép 3 cột description, diagnostic_test, treatment
    và đặt nhãn tiêu đề trước mỗi đoạn.
    """
    # Đọc & strip từng cột
    desc  = str(row.get("description", "")).strip()
    diag  = str(row.get("diagnostic_test", "")).strip()
    treat = str(row.get("treatment", "")).strip()

    parts = []
    if desc and desc.lower() != "nan":
        parts.append("**Description:**\n" + desc)
    if diag and diag.lower() != "nan":
        parts.append("**Diagnostic Tests:**\n" + diag)
    if treat and treat.lower() != "nan":
        parts.append("**Treatment:**\n" + treat)

    return "\n\n".join(parts)

# Tạo cột merged_text
df["merged_text"] = df.apply(merge_row, axis=1)



PROMPT_TEMPLATE = r"""
You are an expert medical knowledge extractor and technical writer.

**Task**

From the unstructured input text delimited by <SOURCE> … </SOURCE>, produce a **fully‑formatted Markdown document** that follows exactly the layout below.  
Include **only** facts that are explicitly stated or clearly implied in the source and that relate to the **same disease**.  
If a required item is missing, write “Information not available”.  
Use concise, fact‑based English (U.S. spelling).

---

### Comprehensive Overview of **<DISEASE_NAME>**  
*(Clear structure, English language)*  

---

## 1. Overview & Molecular Features  
- **Subtype:** …  
- **Genetic feature:** …  
- **Histology:** …  
- **Prevalence:** …  

---

## 2. Common Signs & Symptoms  
> Not every patient shows all manifestations below; list only those mentioned or clearly implied in the source.  
> If a symptom appears more than once with similar wording, list it only once using the most informative phrasing.

- …
- …
- …

---

## 3. Diagnostics  

*(Add one bullet block per diagnostic method mentioned.)*  

- **Method:** …  
  - **Main role:** …  
  - **Notes:** …  

---

## 4. Standard Therapy  

### 4.1 First‑line regimen  
- …

### 4.2 Salvage / second‑line  
- …

### 4.3 Individualised care  
- …

---

## 5. Additional Key Points  
1. …  
2. …  
3. …  

---

## 6. Principal References  

List up to five source‑style citations:  

- **[1]** … – …  
- **[2]** … – …  
- **[3]** … – …  

---

> **Disclaimer:** The information above is for reference only. Diagnosis, prognosis, and treatment must be determined by a qualified medical professional.
In the References section, omit rows whose highlight would be “Information not available”.

---

### Strict Formatting & Content Rules for the LLM

1. Extract facts **only** from the text inside `<SOURCE> … </SOURCE>`; do **not** invent or reuse generic content.  
2. Preserve every heading, bullet, and block quote **exactly** as shown, replacing only the ellipses ( … ) with data. Do **not** introduce tables or other formats.  
3. Do **not** include the raw source text in the output.  
4. Replace `<DISEASE_NAME>` with the disease name found (or inferred) in the source.  
5. Keep all numbers (dates, percentages) exactly as in the source.  
6. In the References section, list up to five citations that appear in the source. Use short publication‑style labels such as “Vodicka 2022”, “NIH 2024” – **never** generic phrases like “Exams and Tests”.  
7. Exclude any marketing, legal, or unrelated commentary present in the source.

---

<SOURCE>
[[SOURCE_TEXT]]
</SOURCE>
""".strip()

```

```python
# ========= TÙY CHỈNH =========
DEBUG_PROMPT = False
TEST_MODE    = False
CHKP_EVERY   = 50
START_GID    = 2470               # ← chạy tiếp từ đây
# =============================

# --- Resume: lấy các global_index đã xong ---
done_idx = set()
if RESUME_RUN and Path(OUTPUT_JSONL).exists():
    with open(OUTPUT_JSONL, "r", encoding="utf-8") as f:
        for line in f:
            try:
                obj = json.loads(line)
                gid = obj.get("global_index", obj.get("row_index"))
                if gid is not None:
                    done_idx.add(int(gid))
            except (json.JSONDecodeError, TypeError, ValueError):
                continue

# --- Cắt DataFrame cho gọn ---
if "global_index" in df.columns:
    df_sub = df[df["global_index"] >= START_GID].reset_index(drop=True)
else:
    df_sub = df.iloc[START_GID:].reset_index(drop=True)

out_f      = open(OUTPUT_JSONL, "a", encoding="utf-8")
errors     = 0
error_rows = []
new_rows   = 0

for offset, row in tqdm(df_sub.iterrows(),
                        total=len(df_sub),
                        desc="Gemini progress"):
    # ----- Tính global_index -----
    if "global_index" in row:
        gid = int(row["global_index"])
    else:
        gid = START_GID + offset

    if gid in done_idx:            # skip đã làm
        continue

    # ---------- Build prompt ----------
    prompt = (PROMPT_TEMPLATE
              .replace("<DISEASE_NAME>", row.get("name", "Unknown"))
              .replace("[[SOURCE_TEXT]]", row["merged_text"]))

    # ---------- Call Gemini ----------
    try:
        resp     = genai_client.models.generate_content(model=MODEL_NAME,
                                                        contents=prompt)
        markdown = resp.text
        status   = "ok"
    except Exception as e:
        markdown = str(e)
        status   = "error"
        errors  += 1
        error_rows.append((gid, row.get("name", ""), str(e)[:200]))
        print(f"\n❌ Error at global_index {gid} ({row.get('name','')}): {e}\n")

    # ---------- Ghi record ----------
    record = {
        "global_index": gid,
        "name"       : row.get("name", ""),
        "markdown"   : markdown,
        "status"     : status
    }
    if DEBUG_PROMPT:
        record["prompt_text"] = prompt

    out_f.write(json.dumps(record, ensure_ascii=False) + "\n")
    new_rows += 1

    if new_rows % CHKP_EVERY == 0:
        out_f.flush()

    time.sleep(SLEEP_SEC)
    break

out_f.close()

# ---------- Tổng kết ----------
print(f"✅ Finished | New rows: {new_rows} | Errors: {errors}")
if errors:
    print("\nDanh sách lỗi:")
    for gid, name, msg in error_rows:
        print(f"- global_index {gid} | {name} | {msg}")
```
